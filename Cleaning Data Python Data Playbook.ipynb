```python
import pandas as pd
```

```python
data = pd.read_csv('Realestate.csv', low_memory=False)
```

```python
data
```

```python
data.dtypes
```

```python
houseAge = data.houseAge
```

```python
#normalize the data in the houseAge column
norm = (houseAge - houseAge.mean()) / houseAge.std()
norm
```

```python
#Create a new column labeled houseAgeStandardized to hold the data that was normalized
data['houseAgeStandardized'] = norm
```

```python
data
```

```python
# drop numberOfConvenienceStores column and apply this change to the data set. 
data.drop(columns=['numberOfConvenienceStores'], inplace=True) 
```

```python
data
```

```python
#change the transaction column name to transactionDate
data.rename(columns={'transaction':'transactionDate'}, inplace=True)
```

```python
# find and remove any duplicate rows.
data.loc[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True)
```

```python
# Find any missing values and fill them with the mean
data.fillna(value={'housePriceOfUnitArea': data.housePriceOfUnitArea.mean()}, inplace=True)
```
